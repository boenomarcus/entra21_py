<b>Exercício 01</b>

Crie um programa que cadastre 5 clientes. 

Cada cliente possui: Nome, sexo, Telefone<br>
(Guarde estes dados em listas separadas).

Depois mostre os dados cadastrados no seguinte formato:



\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*<br>
Relatório de clientes cadastrados <br>
\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*<br>
Nome: <br>
Sexo:<br>
Telefone:<br>
\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*\*<br>

```
 
# Apresentação
print("\n" + "=-"*25)
print(f"{'CADASTRO DE CLIENTES':^50}")
print("=-"*25)

# Criando listas para clientes
nomes = []
sexos = []
telefones = []

# Captando input do usuário
for i in range(5):
    print(f"\n--- Cadastro do cliente {i+1}")
    nomes.append(input("Nome: "))
    sexos.append(input("Sexo [M/F]: ").strip().upper()[0])
    telefones.append(input("Telefone: "))

# Apresentando informações cadastradas
print("\n\n" + "*"*50)
print(f"{'CLIENTES CADASTRADOS':^50}")
print("*"*50)
for i in range(len(nomes)):
    print(f"\n---- Cliente {i+1}:")
    print(f"Nome: {nomes[i]}")
    print(f"Sexo: {sexos[i]}")
    print(f"Telefone: {telefones[i]}")
print("*"*50 + "\n")
```

<b>Exercício 02</b>

O id de um cliente é um código único (só aquela pessoa tem) composto por números inteiros que inicia do número 1 e vai aumentando de 1 em 1 enquanto for necessário.

Exemplo:<br>
id: 1<br>
Nome: Dudu<br>

id: 2<br>
Nome: Marta<br>

id: 3<br>
Nome: Pedro<br>

<br>
<b>ATENÇÃO!!!!<br>
O id é um número atribuido automáticamente! O cliente não escolhe o número. O seu programa deve fazer o cadastro deste id automaticamente.</b><br><br><br>


Com isso, crie um cadastro de clientes que receba o id, nome e idade. Depois mostre os dados dos clientes individualmente.
(cadastre no minimo 4 clientes)


```
 
# Apresentação
print("\n" + "=-"*25)
print(f"{'CADASTRO DE CLIENTES':^50}")
print("=-"*25)

# Criando listas para clientes
ids = []
nomes = []
idades = []

# Captando input do usuário
for i in range(4):
    print(f"\n--- Cadastro do cliente {i+1}")
    ids.append(i+1)
    nomes.append(input("Nome: "))
    idades.append(int(input("Idade: ")))

# Apresentando informações cadastradas
print("\n\n" + "*"*50)
print(f"{'CLIENTES CADASTRADOS':^50}")
print("*"*50)
for i in range(len(ids)):
    print(f"\n---- Cliente {ids[i]}:")
    print(f"Id: {ids[i]}")
    print(f"Nome: {nomes[i]}")
    print(f"Idade: {idades[i]} anos")
print("*"*50)
```

<b>Exercício 03</b>

3.1) Crie um programa que cadastre o id, nome, sexo e a idade de 5 clientes.

3.2) Depois mostre os dados dos 5 clientes.

3.3) Peça para que o usuário escolha um id de um cliente e mostre as informações deste cliente.

3.4) Crie um filtro que ao digitar um id de um cliente mostre as seguintes informações:<br>
- Para menores de 17 anos: Entrada Proibida
- Para maiores de 17 anos: Entrada Liberada
- Para o sexo Feminino: 50% de desconto
- Para o sexo Masculino: 5% de desconto



```
 
# Apresentação
print("\n" + "=-"*25)
print(f"{'CADASTRO DE CLIENTES':^50}")
print("=-"*25)

# Criando lista para armazenar info
dados = [[], [], [], []]

# Captando input do usuário (Ex 3.1)
for i in range(5):
    print(f"\n--- Cadastro do cliente {i+1}")
    dados[0].append(i+1)
    dados[1].append(input("Nome: "))
    dados[2].append(input("Sexo [M/F]: ").strip().upper()[0])
    dados[3].append(int(input("Idade: ")))

# Apresentando clientes cadastrados (Ex. 3.2)
print("\n\n" + "*"*50)
print(f"{'CLIENTES CADASTRADOS':^50}")
print("*"*50)
for i in range(len(dados[0])):
    # Captando info do cliente
    id, nome, sexo, idade = dados[0][i], dados[1][i], dados[2][i], dados[3][i] 
    # Apresentando em tela
    print(f"ID: {id}, Nome: {nome}, Sexo: {sexo}, Idade: {idade}")
print("*"*50 + "\n")

# Captando novo input do usuário (Ex. 3.3)
id = int(input("\nDeseja ver informações de qual cliente (Digite o ID)?! "))

# Apresentando info (Ex. 3.4)
print(f"\n> Você escolheu o cliente '{dados[1][id-1]}' [ID: {dados[0][id-1]}]")
#
if dados[3][id-1] >= 17:
    print(f"   - Entrada LIBERADA ({dados[1][id-1]} tem 17+ anos)!")
else:
    print(f"   - Entrada PROIBIDA ({dados[1][id-1]} tem menos de 17 anos)!")
#
if dados[2][id-1] == 'F':
    print(f"   - Desconto de 50% (Sexo Feminino)!\n")
else:
    print(f"   - Desconto de 5% (Sexo Masculino)!\n")
```
